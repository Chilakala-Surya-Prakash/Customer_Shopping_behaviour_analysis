## 📌 Notebook Overview
This notebook performs **Exploratory Data Analysis (EDA)** and **Data Cleaning** on retail customer shopping behavior data. The goal is to clean feature columns, inspect customer demographic distributions, evaluate spending habits, and prepare data for SQL business querying and Power BI dashboard visualization.

### 1. Setup Libraries & Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plotting style
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
# Ingest dataset
df = pd.read_csv('../data/customer_shopping_behavior.csv')
df.head()
```

### 2. Data Inspection & Cleaning

```python
# Check dataset shape and schema
print(f"Dataset Shape: {df.shape}")
df.info()
```

```python
# Check missing values and duplicate records
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nDuplicate rows: {df.duplicated().sum()}")
```

```python
# Standardize column names to snake_case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.head()
```

### 3. Summary Statistics & Univariate EDA

```python
# Summary metrics for numerical features
df.describe()
```

```python
# Distribution of Purchase Amount
plt.figure(figsize=(10, 5))
sns.histplot(df['purchase_amount_usd'], kde=True, color='teal', bins=30)
plt.title('Distribution of Purchase Amount ($USD)')
plt.xlabel('Purchase Amount ($USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Customer Distribution by Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender', palette='Set2')
plt.title('Customer Breakdown by Gender')
plt.xlabel('Gender')
plt.ylabel('Customer Count')
plt.show()
```

### 4. Bivariate Analysis & Export

```python
# Spend comparison by Subscription Status
df.groupby('subscription_status')['purchase_amount_usd'].agg(['count', 'mean', 'sum']).reset_index()
```

```python
# Save cleaned dataset
df.to_csv('../data/cleaned_customer_shopping_behavior.csv', index=False)
print("Cleaned dataset saved successfully!")
```
